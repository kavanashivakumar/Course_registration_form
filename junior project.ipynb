```python
#faculty login form

def faculty_login(root):
    clear_frame(root)

    label = tk.Label(root, text="Faculty Login Details", font=("Arial", 14))
    label.pack(pady=10)

    tk.Label(root, text="ID").pack()
    ID_entry = tk.Entry(root)
    ID_entry.pack()

    tk.Label(root, text="Password").pack()
    password_entry = tk.Entry(root)
    password_entry.pack()

    next_button = tk.Button(root, text="Login", command=lambda: login_validation(root,ID_entry,password_entry))
    next_button.pack(pady=10)
    
```

# student design

```python
#student login form

def student_login(root):
    clear_frame(root)

    label = tk.Label(root, text="Student Login Details", font=("Arial", 14))
    label.pack(pady=10)

    tk.Label(root, text="USN").pack()
    usn_entry = tk.Entry(root)
    usn_entry.pack()

    tk.Label(root, text="Password").pack()
    password_entry = tk.Entry(root,show="*")
    password_entry.pack()

    next_button = tk.Button(root, text="Login", command=lambda: login(root,usn_entry,password_entry))
    next_button.pack(pady=10)
```

```python
#login validation 
    
def login(root,usn_entry,password_entry):
    usn = usn_entry.get()
    password = password_entry.get()
    
        #querie execution
    con.execute(f"select usn from student_details where usn = '{usn}'")
    data=con.fetchall()#fetching all data according to querie
    
    if data !=[]:#enters only if the data is present in the database
        for x in data:
            for i in x:
                db_usn=i
                break
        # Check login credentials
        if usn.startswith(db_usn) and password == db_usn:
            con.execute(f"select branch from student_details where usn = '{usn}'")
            branch_data = con.fetchall()#fetching all data according to querie
            for y in branch_data:
                for i in y:
                    b_val=i
                    break
            select_semester(root, b_val,usn)
        else:
            messagebox.showerror("Invalid credentials", "Please Enter the Valid USN and Password.")
    else:
        messagebox.showerror("Invalid credentials", "Please Enter the Valid USN and Password.")
```

```python
#selecting the semester
    
def select_semester(root, branch_var,usn):
    clear_frame(root)

    label = tk.Label(root, text="Select Semester", font=("Arial", 14))
    label.pack(pady=10)

    semesters = ["3", "4","5",'6','7','8']
    semester_var = tk.StringVar()
    semester_var.set(semesters[0])

    for semester in semesters:
        semester_radio = tk.Radiobutton(root, text=semester, variable=semester_var, value=semester)
        semester_radio.pack()

    next_button = tk.Button(root, text="Next", command=lambda: enter_details(root, branch_var, semester_var,usn))
    next_button.pack(pady=10)
```

```python
#entering the fees receipt number
def enter_details(root, branch_var, semester_var,usn):
    clear_frame(root)

    label = tk.Label(root, text="Enter Details", font=("Arial", 14))
    label.pack(pady=10)
    
    con.execute(f"select name from student_details where usn = '{usn}'")
    stu_data = con.fetchall()#fetching all data according to querie
    for y in stu_data:
        for i in y:
            name_entry=i
            break
    tk.Label(root, text=f"Name:{name_entry}").pack()
   

    tk.Label(root, text="Fees Receipt Number:").pack()
    receipt_entry = tk.Entry(root)
    receipt_entry.pack()

    next_button = tk.Button(root, text="Next", command=lambda: checking_fee_data(root, branch_var, semester_var, name_entry, receipt_entry,usn))
    next_button.pack(pady=10)
```

```python
#validation for input data of fee recipet number

def checking_fee_data(root, branch_var, semester_var, name_entry, receipt_entry,usn):
    recipt=receipt_entry.get()
    if recipt == "":
        messagebox.showerror("Invalid credentials", "Please Enter the Fees Receipt Number.")
    else:
        select_courses(root, branch_var, semester_var, name_entry, receipt_entry,usn)
```

```python
#validation for the checkbox of courses

def validate_and_register(root, course_vars, name, receipt_number,usn,semester):
    selected_courses = [course[0] for course in course_vars if course[1].get() == 1]

    if not selected_courses:
        messagebox.showerror("Error", "Please select at least one course.")
    else:
        registration_success(root, name, receipt_number, selected_courses,usn,semester)
```

```python
#CSE courses
def branch_cse(root,semester):
    if semester == "3":
        courses = ["Maths", "Logic Design", "DSA",'AI']
    elif semester == "4":
        courses = ["DBMS", "Maths", "IoT", "Java"]
    elif semester == "5":
        courses = ["C++", "DSA", "Maths", "English"]
    elif semester == "6":
        courses = ["Maths", "Python", "UHV", "Node JS"]
    elif semester == "7":
        courses = ["C++", "DSA", "Maths", "English"]
    elif semester == "8":
        courses = ['Project Final Phase', "Internship"]
    
    return courses    
    
```

```python
#ISE courses
def branch_ise(root,semester):
    if semester == "3":
        courses = ["Maths", "Design", "DAA","C++"]
    elif semester == "4":
        courses = ["DBMS", "Maths", "IoT", "Java"]
    elif semester == "5":
        courses = ["C++", "DSA", "Placement", "English"]
    elif semester == "6":
        courses = ["Maths", "Python", "UHV", "Node JS"]
    elif semester == "7":
        courses = ["C#", "TOC", "Advance Java", "Kannada"]
    elif semester == "8":
        courses = ['Project Final Phase', "Internship"]
    
    return courses 
```

```python
#select the course list

def select_courses(root, branch, semester_var, name_entry, receipt_entry,usn):
    name = name_entry
    receipt_number = receipt_entry.get()

    clear_frame(root)

    label = tk.Label(root, text="Select Courses", font=("Arial", 14))
    label.pack(pady=10)
    semester = semester_var.get()

    if branch == "CSE":
        
        courses = branch_cse(root,semester)
    elif branch == "ISE":
        
        courses = branch_ise(root,semester)    
        
    else:
        clear_frame(root)
        messagebox.showerror("Error", "Please contect Department HOD")
        label = tk.Label(root, text="You can close the window!", font=("Arial", 10))
        label.pack(pady=10)
        

    course_vars = []
    for course in courses:
        var = tk.IntVar()
        course_check = tk.Checkbutton(root, text=course, variable=var)
        course_check.pack()
        course_vars.append((course, var))

    next_button = tk.Button(root, text="Register", command=lambda: validate_and_register(root, course_vars, name, receipt_number,usn,semester))
    next_button.pack(pady=10)
```

```python
#final display of the details

def registration_success(root, name, receipt_number, selected_courses,usn,semester):
    clear_frame(root)

    label = tk.Label(root, text="Registration Successful!", font=("Arial", 14))
    label.pack(pady=10)

    label_name = tk.Label(root, text=f"Name: {name}")
    label_name.pack()
    label_name = tk.Label(root, text=f"USN: {usn}")
    label_name.pack()
    label_receipt = tk.Label(root, text=f"semester: {semester}")
    label_receipt.pack()
    label_receipt = tk.Label(root, text=f"Receipt Number: {receipt_number}")
    label_receipt.pack()

    label_courses = tk.Label(root, text="Selected Courses:")
    label_courses.pack()
    count = 1

    course_1='null'
    course_2='null'
    course_3='null'
    course_4='null'
    for course in selected_courses:
        label_course = tk.Label(root, text=f"{count}.{course}")
        label_course.pack()
        if count == 1:
            course_1=course
        elif count == 2:
            course_2=course
        elif count == 3:
            course_3=course
        elif count == 4:
            course_4=course
        count+=1
        
    #entering the data to the database
    con.execute(f"insert into data_entry(usn,sem,receipt_no, course_1, course_2,course_3,course_4) values('{usn}',{semester},'{receipt_number}','{course_1}','{course_2}','{course_3}','{course_4}')")
    db.commit() 
      
    label = tk.Label(root, text="NOTE: Submit your fee recepit xerox to the Mentors")
    label.pack(pady=10)
```

```python
# main code starting

if __name__ == "__main__":
    root = tk.Tk()
    root.title("Online Course Registration")
    root.geometry("1000x600")

    faculty_button = tk.Button(root, text="Login as Faculty", command=lambda: faculty_login(root))
    faculty_button.pack(pady=5)

    student_button = tk.Button(root, text="Login as Student", command=lambda: student_login(root))
    student_button.pack(pady=5)

    root.mainloop()
```
